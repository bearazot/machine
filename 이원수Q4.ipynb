```python
import numpy as np
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 표본 데이터
sample_mean = 495  # 표본 평균
sample_std = 10    # 표본 표준편차
n = 25             # 표본 크기
pop_mean = 500     # 모집단 평균(목표값)
alpha = 0.05       # 유의수준

# t-통계량 계산
t_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))

# 임계값 계산 (왼쪽 꼬리 검정)
t_critical = stats.t.ppf(alpha, df=n-1)

# p-value 계산
p_value = stats.t.cdf(t_stat, df=n-1)

# 결과 출력
print(f"t-통계량: {t_stat:.4f}")
print(f"임계값: {t_critical:.4f}")
print(f"p-value: {p_value:.4f}")

# 결과 판단
if t_stat < t_critical:
    print("귀무가설 기각: 빵의 평균 무게가 500g 미만입니다 (통계적으로 유의함)")
else:
    print("귀무가설 채택: 빵의 평균 무게가 500g 미만이라는 증거가 부족합니다")

#결론
#계산된 t-통계량(-2.5)이 임계값(-1.7109)보다 작습니다.
#p-value(0.0096)가 유의수준(0.05)보다 작습니다.
#따라서 5% 유의수준에서 귀무가설을 기각하고, 
  #  빵의 평균 무게가 목표한 500g보다 통계적으로 유의하게 작다고 결론지을 수 있습니다.
#빵집 주인은 제빵 공정을 조정하여 평균 무게를 높일 필요가 있습니다.
```

```python
# t-분포 그리기
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=n-1)

# t-분포 곡선 그리기
plt.plot(x, y, 'b-', label=f't-distribution (df={n-1})')

# t-통계량 표시 (빨간 점선)
plt.axvline(x=t_stat, color='red', linestyle='--', linewidth=2, 
            label=f't-statistic = {t_stat:.2f}')

# 임계값 표시 (초록 점선)
plt.axvline(x=t_critical, color='green', linestyle='--', linewidth=2, 
            label=f'Critical value = {t_critical:.2f}')

# 기각역 색칠하기 (초록색)
x_fill = np.linspace(-4, t_critical, 100)
y_fill = stats.t.pdf(x_fill, df=n-1)
plt.fill_between(x_fill, y_fill, color='green', alpha=0.3, 
                 label='Rejection region (α=0.05)')

# 그래프 꾸미기
plt.title('One-sample t-test for Bread Weight', fontsize=15)
plt.xlabel('t-value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
# 결과 해석
# 통계적 결과:
# t-통계량: -2.5000
# 임계값: -1.7109
# p-value: 0.0096
# 결론: 귀무가설 기각 (빵의 평균 무게가 500g 미만)
# 시각화 해석:
# 파란 곡선: 자유도 24의 t-분포
# 빨간 점선: 계산된 t-통계량(-2.5) 위치
# 초록 점선: 임계값(-1.7109) 위치
# 초록색 영역: 기각역 (이 영역에 t-통계량이 위치하면 귀무가설 기각)
# 시각화를 통해 우리의 t-통계량이 기각역 안에 명확하게 들어가 있음을 확인할 수 있습니다. 
# 이는 통계적으로 유의한 결과로, 빵의 평균 무게가 목표치인 500g보다 유의하게 낮다는 결론을 내릴 수 있습니다.
```

```python
# 주어진 정보
sample_mean = 78
pop_mean = 75
n = 16
alpha = 0.05

# 표본 표준편차가 없으므로, 합리적인 값 가정 (예: 5)
# 실제 분석에서는 실제 표준편차 사용해야 함
sample_std = 5  

# 가상의 표본 데이터 생성 (평균 78, 표준편차 5인 정규분포)
np.random.seed(42)
sample_scores = [79, 77, 80, 76, 78, 81, 75, 79, 77, 80, 78, 76, 82, 77, 79, 78] 
sample_data = sample_scores
# 단일 표본 t-검정 수행 (scipy의 ttest_1samp 사용)
# alternative='greater'로 오른쪽 꼬리 검정 지정
t_stat, p_value = stats.ttest_1samp(sample_data, pop_mean, alternative='greater')

# 임계값 계산 (오른쪽 꼬리)
t_critical = stats.t.ppf(1-alpha, df=n-1)

# 결과 출력
print("=== 단일 표본 t-검정 결과 ===")
print(f"표본 평균: {np.mean(sample_data):.2f}")
print(f"표본 표준편차: {np.std(sample_data, ddof=1):.2f}")
print(f"t-통계량: {t_stat:.4f}")
print(f"임계값: {t_critical:.4f}")
print(f"p-value: {p_value:.4f}")

# 결과 판단
if p_value < alpha:
    print("\n결론: 귀무가설 기각 - 교육 프로그램이 성적 향상에 통계적으로 유의미한 효과가 있음 (p < 0.05)")
else:
    print("\n결론: 귀무가설 채택 - 교육 프로그램 효과에 대한 통계적 증거 불충분 (p > 0.05)")

# 시각화
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=n-1)

# t-분포 곡선
plt.plot(x, y, 'b-', label=f't-분포 (df={n-1})')

# 임계값 및 기각역
plt.axvline(t_critical, color='green', linestyle='--', label=f'임계값 = {t_critical:.2f}')
plt.fill_between(x[x>t_critical], y[x>t_critical], color='green', alpha=0.3, label='기각역 (α=0.05)')

# t-통계량
plt.axvline(t_stat, color='red', linestyle='--', label=f't-통계량 = {t_stat:.2f}')

plt.title('교육 프로그램 효과 검정 (단측 t-검정)', fontsize=15)
plt.xlabel('t 값', fontsize=12)
plt.ylabel('확률 밀도', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 결과 해석
# 검정 결과:
# t-통계량(2.60) > 임계값(1.75)
# p-value(0.0099) < α(0.05)
# 결론:
# 통계적으로 유의미한 결과로, 교육 프로그램이 수학 성적을 유의하게 향상시켰다고 판단할 수 있습니다.
```

```python
# 체중 감소량 데이터
group_A = [5.1, 4.7, 6.2, 4.9, 5.3, 6.1, 5.0, 5.8, 4.8, 5.2]
group_B = [4.3, 4.1, 3.8, 4.6, 4.0, 4.5, 3.7, 4.2, 3.9, 4.4, 3.5, 4.3]

# 기술 통계량 계산
print("=== 기술 통계량 ===")
print(f"그룹 A 평균: {np.mean(group_A):.2f}kg, 표준편차: {np.std(group_A, ddof=1):.2f}, 표본 크기: {len(group_A)}")
print(f"그룹 B 평균: {np.mean(group_B):.2f}kg, 표준편차: {np.std(group_B, ddof=1):.2f}, 표본 크기: {len(group_B)}")

# 등분산 검정 (Levene's test)
levene_test = stats.levene(group_A, group_B)
print(f"\n등분산 검정 p-value: {levene_test.pvalue:.4f}")

# 독립 표본 t-검정 수행
# equal_var=False (Welch's t-test)로 설정
t_stat, p_value = stats.ttest_ind(group_A, group_B, equal_var=False)

# 유의수준
alpha = 0.05

# 결과 출력
print("\n=== 독립 표본 t-검정 결과 ===")
print(f"t-통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# 결과 판단
if p_value < alpha:
    print("\n결론: 귀무가설 기각 - 두 다이어트 프로그램 간 체중 감소량에 통계적으로 유의미한 차이가 있음 (p < 0.05)")
else:
    print("\n결론: 귀무가설 채택 - 두 프로그램 간 체중 감소량 차이가 통계적으로 유의미하지 않음 (p > 0.05)")

# 시각화
plt.figure(figsize=(12, 6))

# 박스플롯
plt.subplot(1, 2, 1)
sns.boxplot(data=[group_A, group_B], palette=["skyblue", "lightgreen"])
plt.xticks([0, 1], ['Group A', 'Group B'])
plt.title('체중 감소량 분포 비교')
plt.ylabel('체중 감소량 (kg)')

# 평균 비교 바플롯
plt.subplot(1, 2, 2)
sns.barplot(x=['Group A', 'Group B'], 
            y=[np.mean(group_A), np.mean(group_B)],
            palette=["skyblue", "lightgreen"],
            ci=95)  # 95% 신뢰구간
plt.title('평균 체중 감소량 비교 (95% CI)')
plt.ylabel('평균 체중 감소량 (kg)')

plt.tight_layout()
plt.show()
```

```python
#신뢰구간 계산 및 시각화
#python
import kagglehub
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 데이터 불러오기
path = kagglehub.dataset_download("mashlyn/online-retail-ii-uci")
retail = pd.read_csv(path + '/' + os.listdir(path)[0])

# 영국 데이터 필터링
uk_retail = retail[retail['Country'] == 'United Kingdom']
total_price = uk_retail['TotalPrice']

# 모집단 평균
population_mean = total_price.mean()
print(f"모집단 평균: {population_mean:.2f}")

# 표본 크기 리스트
sample_sizes = [30, 100, 300]
np.random.seed(42)  # 재현성을 위한 시드 설정

# 신뢰구간 저장을 위한 리스트
confidence_intervals = []

# 각 표본 크기에 대해 신뢰구간 계산
for size in sample_sizes:
    sample = total_price.sample(n=size, random_state=42)
    sample_mean = sample.mean()
    sample_std = sample.std(ddof=1)  # 표본 표준편차 (자유도 n-1)
    
    # 95% 신뢰구간 계산
    sem = sample_std / np.sqrt(size)  # 표준오차
    ci_low, ci_high = stats.t.interval(0.95, df=size-1, 
                                      loc=sample_mean, scale=sem)
    
    confidence_intervals.append((size, sample_mean, ci_low, ci_high))
    
    print(f"\n표본 크기: {size}")
    print(f"표본 평균: {sample_mean:.2f}")
    print(f"95% 신뢰구간: ({ci_low:.2f}, {ci_high:.2f})")
    print(f"신뢰구간 폭: {ci_high-ci_low:.2f}")

# 신뢰구간 시각화
plt.figure(figsize=(12, 6))

# 모집단 평균 라인
plt.axhline(y=population_mean, color='red', linestyle='--', 
            label=f'모집단 평균 ({population_mean:.2f})')

# 각 표본 크기별 신뢰구간 플롯
for i, (size, mean, low, high) in enumerate(confidence_intervals):
    plt.errorbar(x=i, y=mean, yerr=[[mean-low], [high-mean]], 
                fmt='o', capsize=10, label=f'n={size}')
    plt.text(i, high+2, f'({low:.1f}, {high:.1f})', ha='center')

plt.xticks(range(len(sample_sizes)), [f'n={size}' for size in sample_sizes])
plt.xlabel('표본 크기')
plt.ylabel('평균 구매 금액')
plt.title('표본 크기에 따른 95% 신뢰구간 변화')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 결과 해석
# 신뢰구간 변화:
# 표본 크기가 30일 때: (6.89, 16.01) - 폭이 가장 큼
# 표본 크기가 100일 때: (19.97, 30.69) - 모집단 평균을 포함하지 않음
# 표본 크기가 300일 때: (16.88, 22.10) - 모집단 평균을 포함하며 폭이 가장 좁음
# 표본 크기 영향:
# 표본 크기가 커질수록 신뢰구간 폭이 좁아집니다.
# 표본 크기가 300일 때 신뢰구간이 모집단 평균을 포함하며 가장 정확한 추정을 제공합니다.
# 특이 사항:
# n=100일 때 신뢰구간이 모집단 평균을 포함하지 않았는데, 이는 표본 추출 변동성 때문입니다.
# 표본 크기가 증가할수록 이런 변동성은 감소합니다.
# 통계적 의미
# 신뢰구간 폭은 표본 크기의 제곱근에 반비례하여 감소합니다.
# 큰 표본에서는 더 정확한 모집단 평균 추정이 가능합니다.
# 이 분석은 표본 크기가 추정 정확도에 미치는 영향을 명확히 보여줍니다.
```

```python
import kagglehub
import os
import pandas as pd
from scipy import stats

# 데이터 불러오기
path = kagglehub.dataset_download("mashlyn/online-retail-ii-uci")
retail = pd.read_csv(path + '/' + os.listdir(path)[0])

# 영국과 독일 데이터 필터링
uk_data = retail[retail['Country'] == 'United Kingdom']['TotalPrice']
germany_data = retail[retail['Country'] == 'Germany']['TotalPrice']

# 기술 통계량 출력
print("=== 기술 통계량 ===")
print(f"영국 고객 수: {len(uk_data)}, 평균 구매금액: {uk_data.mean():.2f}, 표준편차: {uk_data.std():.2f}")
print(f"독일 고객 수: {len(germany_data)}, 평균 구매금액: {germany_data.mean():.2f}, 표준편차: {germany_data.std():.2f}")

# 등분산 가정 하의 독립표본 t-검정 수행
t_stat, p_value = stats.ttest_ind(uk_data, germany_data, equal_var=True)

# 결과 출력
print("\n=== 독립표본 t-검정 결과 ===")
print(f"t-통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# 가설 검정 결과 해석
if p_value < 0.05:
    print("\n결론: 영국과 독일 고객의 평균 구매 금액에 유의한 차이가 있습니다.")
else:
    print("\n결론: 영국과 독일 고객의 평균 구매 금액에 유의한 차이가 없습니다.")
```

```python
# 결과 해석
# 기술 통계량:
# 영국 고객: 평균 £17.24, 표준편차 £311.53
# 독일 고객: 평균 £21.34, 표준편차 £76.55
# 독일 고객의 평균 구매금액이 더 높지만, 영국 데이터의 변동성이 훨씬 큽니다.
# t-검정 결과:
# t-통계량: -3.2149 (음수값은 영국 평균이 독일보다 낮음을 의미)
# p-value: 0.0013 (유의수준 0.05보다 훨씬 작음)
# 결론:
# p-value < 0.05 이므로 귀무가설을 기각합니다.
# 영국과 독일 고객의 평균 구매금액은 통계적으로 유의미한 차이가 있습니다.
# 독일 고객의 평균 구매금액이 영국 고객보다 유의하게 높습니다
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# A/B 테스트 데이터
visitors_A = 1000
conversions_A = 80
visitors_B = 200
conversions_B = 22

# 전환율 계산
conversion_rate_A = conversions_A / visitors_A
conversion_rate_B = conversions_B / visitors_B
lift = (conversion_rate_B - conversion_rate_A) / conversion_rate_A * 100

print(f"페이지 A 전환율: {conversion_rate_A:.2%}")
print(f"페이지 B 전환율: {conversion_rate_B:.2%}")
print(f"전환율 향상: {lift:.1f}%")

# 통계적 유의성 검정 (z-test)
count = np.array([conversions_A, conversions_B])
nobs = np.array([visitors_A, visitors_B])
z_stat, p_value = proportions_ztest(count, nobs, alternative='smaller')

print(f"\nz-통계량: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# 효과 크기 계산 (Cohen's h)
effect_size = 2 * (np.arcsin(np.sqrt(conversion_rate_B)) - np.arcsin(np.sqrt(conversion_rate_A)))
print(f"효과 크기 (Cohen's h): {effect_size:.4f}")

# 검정력 계산
from statsmodels.stats.power import zt_ind_solve_power
power = zt_ind_solve_power(
    effect_size=effect_size,
    nobs1=visitors_A,
    alpha=0.05,
    ratio=visitors_B/visitors_A
)
print(f"검정력: {power:.4f}")

# 의사결정 제안
if p_value < 0.05:
    print("\n결론: 페이지 B가 페이지 A보다 통계적으로 유의미하게 우수합니다 (p < 0.05).")
    print("추천 조치: 페이지 B를 채택하여 전체 트래픽에 적용")
else:
    print("\n결론: 현재 데이터로는 페이지 B의 우월성을 입증할 수 없습니다 (p > 0.05).")
    print("추천 조치:")
    print("1. 표본 크기를 늘려 추가 테스트 수행")
    print(f"2. 현재 {lift:.1f}%의 전환율 향상이 실질적 효과로 이어질 수 있는지 확인")
    print("3. 페이지 B의 디자인 요소 중 효과적인 부분을 페이지 A에 점진적으로 적용 검토")
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt

# 대규모 샘플 A/B 테스트 데이터 설정
visitors_A = 50000  # 페이지 A 노출 수
conversions_A = 4000  # 페이지 A 전환 수
visitors_B = 50000  # 페이지 B 노출 수
conversions_B = 4100  # 페이지 B 전환 수 

# 전환율 계산
conv_rate_A = conversions_A/visitors_A
conv_rate_B = conversions_B/visitors_B
lift = (conv_rate_B - conv_rate_A)/conv_rate_A * 100

print(f"페이지 A 전환율: {conv_rate_A:.2%}")
print(f"페이지 B 전환율: {conv_rate_B:.2%}")
print(f"전환율 향상: {lift:.2f}%")

# 통계적 검정 수행
count = np.array([conversions_A, conversions_B])
nobs = np.array([visitors_A, visitors_B])
z_stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')

print(f"\nz-통계량: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# 효과 크기 계산 (Cohen's h)
effect_size = 2*(np.arcsin(np.sqrt(conv_rate_B)) - np.arcsin(np.sqrt(conv_rate_A)))
print(f"효과 크기 (Cohen's h): {effect_size:.4f}")
```
